```python
import pandas as pd

df = pd.read_csv("sis_summary_results.csv")

import matplotlib.pyplot as plt
# Plot
plt.figure(figsize=(8, 5))
plt.plot(df['activity'], df['SteadyState'], marker='.', linestyle='--', color='black')
plt.xlabel('activity rate')
plt.ylabel('Steady State Infection')
plt.title('Steady State Infection vs. activity rate')
plt.grid(True)
plt.tight_layout()
plt.show()

# df[['m', 'SteadyState']]
df
```
