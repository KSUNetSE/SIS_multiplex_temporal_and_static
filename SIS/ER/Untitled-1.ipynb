```python
import pandas as pd

df = pd.read_csv("sis_summary_results.csv")

import matplotlib.pyplot as plt
# Plot
plt.figure(figsize=(8, 5))
plt.plot(df['m'], df['SteadyState'], marker='o', linestyle='-', color='black')
plt.xlabel('m (Average Degree or Link Count)')
plt.ylabel('Steady State Infection')
plt.title('Steady State Infection vs. m')
plt.grid(True)
plt.tight_layout()
plt.show()

# df[['m', 'SteadyState']]
df
```
