```python
# === Parameters ===
adjlist = '../ER-1000'
static_graph = nx.read_adjlist(f"{adjlist}.txt", nodetype=int)
N = static_graph.number_of_nodes()
static_edges_set = set(static_graph.edges())

m_values = [0, 15]
activity = 0.1
activity_rates = {node: activity for node in static_graph.nodes}

timesteps = 200
num_runs = 100

beta = 0.08
mu = 1

m_list   = np.arange(0, 12, 1)                     # m = 0,2,4,...,40
tau_list = np.round(np.arange(0.15, 0.25, 0.006), 3)  # tau sweep

transient_cutoff = 100
initial_infected_ratio = 0.01

csv_file = "sis_summary_results.csv"


A = nx.to_numpy_array(static_graph, dtype=int)
eigenvalues, _ = np.linalg.eig(A)
lambda_max_A = np.max(eigenvalues)
tau_c = 1 / lambda_max_A
tau = beta/mu
# theoretical m_c (float) and rounded integer m_c for use in simulations
m_c_float = ((1 / tau) - (lambda_max_A)) / (2*activity)
m_c = int(round(m_c_float))

m_c, tau_c, lambda_max_A
```
